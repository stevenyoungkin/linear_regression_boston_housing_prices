```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

## Exploring the Data

```python
print(boston['DESCR'])
```

```python
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df['MEDV'] = boston.target # target data
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

## EDA

```python
sns.displot(df['MEDV'], kde=True, bins=50); # median housing price distribution
```

```python
# Heatmap to show correlation between features
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), annot=True)
plt.tight_layout();
```

```python
# Here we can explore how the housing price data is distributed with respect to other features.
# For example, the average numbers of rooms and the lower status of the population is 
# hihgly correlated with the housing prices.
g = sns.pairplot(df, x_vars=['MEDV'], 
                 y_vars=['RM', 'LSTAT'],
                 height=2, aspect=2, 
                 plot_kws=dict(marker="+", linewidth=1));
```

## Creating and Training a Linear Regression Model

```python
X = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']] # features to train on
y = df['MEDV'] # target values
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression() # Instantiating a LinearRegression object
```

```python
lm.fit(X_train,y_train); # Fit linear model
```

## Model Evaluation

```python
# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

## Model Predictions

```python
predicted = lm.predict(X_test)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, predicted, marker='+')
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show();
```

```python
sns.displot((y_test-predicted),bins=50, kde=True);
```

## Regression Evaluation Metrics

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predicted))
print('MSE:', metrics.mean_squared_error(y_test, predicted))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicted)))
```
